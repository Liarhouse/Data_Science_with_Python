```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rc('font',family='D2Coding')
```

```python
mpl.rcParams['font.family'] ='D2Coding'
```

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 3),
                  index = pd.date_range('2018-1-1', periods = 100),
                  columns = list('ABC'))
df.head(10)
```

```python
df.plot()

plt.title('Pandas의 내장된 plot 메서드 사용 예')
plt.xlabel('날짜')
plt.ylabel('Data')
plt.show()
```

- plot 메서드에 kind 인수를 통해 다양한 plot을 시각화 할 수 있다.
- bar, pie, hist, kde, box, scatter

```python
import seaborn as sns

iris = sns.load_dataset('iris')
ttnc = sns.load_dataset('titanic')
```

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title('붓꽃의 꽃받침의 길이를 이용한 시각화')
plt.xlabel('data')
plt.ylabel('꽃받침의 길이')
plt.show
```

```python
iris[:5].plot.bar(rot=0)
plt.title('Iris 붓꽃 데이터의 막대 그래프')
plt.xlabel('Data')
plt.ylabel('각 Feature의 값')
plt.ylim(0, 7)
plt.legend(loc=1)
plt.show()
```

```python
# 품종별로 각 특성의 평균값을 시각화하시오.
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0)
plt.title('품종별 Feature의 평균')
plt.xlabel('품종')
plt.ylabel('평균')
plt.ylim(0, 7)
plt.legend(loc='best')
plt.show()
```

```python
# 객실등급별 승객수의 비율을 시각화
df3 = ttnc.pclass.value_counts()
df3
```

```python
df3.plot.pie(autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=2, h_pad = 2)
plt.show()
```
